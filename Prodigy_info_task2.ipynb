```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from scipy.stats import skew
```

```python
df=pd.read_csv('amazon.csv')
print(df)
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.dropna(subset=['actual_price'], inplace=True)
```

# Statistics Summary

```python
print(df.describe())
```

# Conversion of  price-related columns to numeric 

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.rstrip('%'), errors='coerce')
```

```python
print(df.describe())
```

# Correlation Analysis

```python
price_vs_rating = df[['actual_price', 'discounted_price']]

correlation = price_vs_rating.corr()

print("Correlation between Actual Price and Discounted Price:\n", correlation)

plt.figure(figsize=(6, 4))
plt.imshow(correlation, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Correlation')
plt.title("Correlation between Actual Price and Discounted Price")
plt.show()
```

## Inference
* A correlation coefficient of 1 implies a perfect positive linear relationship.
* Here, it means that as the 'actual_price' increases, the 'discounted_price' also tends to increase, and vice versa.

# Skewness

```python
columns_to_calculate_skewness = ['rating', 'discounted_price', 'actual_price', 'discount_percentage','rating_count']
df[columns_to_calculate_skewness] = df[columns_to_calculate_skewness].apply(pd.to_numeric, errors='coerce')
skewness_results = df[columns_to_calculate_skewness].apply(lambda x: skew(x.dropna()))
skewness_df = pd.DataFrame({'Column': columns_to_calculate_skewness, 'Skewness': skewness_results})
print(skewness_df)
```


## 1.Rating:

* The skewness for the 'rating' column is approximately -1.24.
* A negative skewness indicates that the distribution is skewed to the left (negatively skewed). In the context of ratings, this suggests that there might be a higher concentration of products with higher ratings.

## 2.Discounted Price:

* The skewness for the 'discounted_price' column is approximately 4.45.
* A positive skewness indicates that the distribution is skewed to the right (positively skewed). This suggests that there might be a tail of higher discounted prices.

## 3.Actual Price:

* The skewness for the 'actual_price' column is approximately 4.56.
* Similar to 'discounted_price', a positive skewness indicates a right-skewed distribution in actual prices.

## 4.Discount Percentage:

* The skewness for the 'discount_percentage' column is approximately -0.29.
* A slight negative skewness suggests a minor leftward skew, but it is close to zero, indicating a relatively symmetric distribution.

## 5.Rating Count:

* The skewness for the 'rating_count' column is approximately 0.62.
* A positive skewness suggests a right-skewed distribution, meaning there might be a tail of products with higher numbers of ratings.
